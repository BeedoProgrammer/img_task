```python
import cv2

image = cv2.imread("UNDERTALE.jpg")
point1 = None
point2 = None

def draw_rectangle(event, x, y, flags, param):
    global point1, point2, crop_img

    if event == cv2.EVENT_LBUTTONDOWN:
        if point1 is None:
            point1 = (x, y)
        elif point2 is None:
            point2 = (x, y)
        
        if point1 and point2:
                x1, y1 = point1
                x2, y2 = point2
                
                x1, x2 = sorted((x1, x2))
                y1, y2 = sorted((y1, y2))
                
                cv2.rectangle(image, (x1, y1), (x2, y2), (255, 255, 255), 1)
                crop_img = image[y1:y2, x1:x2]

cv2.namedWindow('Image')
cv2.setMouseCallback('Image', draw_rectangle)

while True:
    cv2.imshow('Image', image)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

if crop_img is not None:
    cv2.imshow('Cropped Image', crop_img)
    cv2.waitKey(0)
    cv2.imwrite("cropped_image.jpg", crop_img)

cv2.destroyAllWindows()
```

```python
import cv2

image = cv2.imread("UNDERTALE.jpg")
point1 = None
point2 = None

def draw_circle(event, x, y, flags, param):
    global point1, point2

    if event == cv2.EVENT_LBUTTONDOWN:
        if point1 is None:
            point1 = (x, y)
        elif point2 is None:
            point2 = (x, y)
        
        if point1 and point2:
                x1, y1 = point1
                x2, y2 = point2
                
                radius = int(((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5)
                
                cv2.circle(image, (x1, y1), radius, (255, 255, 255), 1)

cv2.namedWindow('Image')
cv2.setMouseCallback('Image', draw_circle)

while True:
    cv2.imshow('Image', image)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

image = cv2.imread("UNDERTALE.jpg")

def extract_color(event, x, y, flags, param):
    global image, lower_bound, upper_bound

    if event == cv2.EVENT_LBUTTONDOWN:
        hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
        pixel_color = hsv_image[y, x]
        
        lower_bound = np.array([
            max(pixel_color[0], 0),
            max(pixel_color[1], 0),
            max(pixel_color[2], 0)
        ], dtype=np.uint8)

        upper_bound = np.array([
            min(pixel_color[0] + 20, 179),
            min(pixel_color[1] + 20, 255),
            min(pixel_color[2] + 20, 255)
        ], dtype=np.uint8)

        mask = cv2.inRange(hsv_image, lower_bound, upper_bound)
        extracted_color = cv2.bitwise_and(image, image, mask = mask)

        cv2.imshow('Color', extracted_color)

cv2.namedWindow('Image')
cv2.setMouseCallback('Image', extract_color)

while True:
    cv2.imshow('Image', image)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
```
